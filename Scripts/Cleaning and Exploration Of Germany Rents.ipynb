```python
df_new = df.drop(['baseRentRange', 'livingSpaceRange', 'yearConstructedRange', 'noRoomsRange' , 'Unnamed: 0.1', 'Unnamed: 0'], axis = 1)
```

```python
# keep the rows that satisfy all three conditions: 
# the maintenance fee is less than 4000, the total rent is less than 6000, 
# and the base rent is less than or equal to the total rent.
df_final = df_new[(df_new['Maintenance_fee'] < 4000) &
            (df_new['Total_Rent'] < 6000) &
            (df_new['Base_rent'] <= df_new['Total_Rent'])]           
```
